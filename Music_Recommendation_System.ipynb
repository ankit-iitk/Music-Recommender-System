```python
x = df.drop(columns = ['track_name','genre'])
```

```python
x
```

```python
from sklearn.preprocessing import StandardScaler

scaler=StandardScaler()
x_scaled = scaler.fit_transform(x)
```

```python
from sklearn.preprocessing import OneHotEncoder
ohe = OneHotEncoder()
genre_encoded = ohe.fit_transform(df[['genre']])
```

```python
genre_encoded = pd.DataFrame(genre_encoded.toarray(),columns=ohe.get_feature_names_out())
```

```python
genre_encoded
```

```python
x_df=pd.DataFrame(x_scaled,columns=x.columns)
```

```python
final_df=pd.concat([x_df,genre_encoded],axis=1)
```

```python
final_df.shape
```

```python
final_df.index = df['track_name']
```

```python
final_df
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_scores = cosine_similarity(final_df)
```

```python
similarity_scores
```

```python
import numpy as np

def recommend_song(song_name):
  index = np.where(final_df.index == song_name)[0][0]
  distances = similarity_scores[index]
  song_list  = sorted(list(enumerate(distances)),key = lambda x:x[1],reverse=True)[1:6]

  for i in song_list:
    print(final_df.index[i[0]])
```
